```python
from pycaret import regression
from pycaret.regression import *
```

```python
train['FirstFloorSF']=train['1stFlrSF']
train['SecondFloorSF']=train['2ndFlrSF']
```

```python
train=train[['LotArea','GarageArea','FirstFloorSF','SecondFloorSF','PavedDrive','SaleCondition','FullBath','HalfBath','SalePrice']]
```

```python
train.head()
```

```python
reg = setup(data = train, 
             target = 'SalePrice',
             numeric_imputation = 'mean',
             categorical_features = ['PavedDrive','SaleCondition'] , 
             #ignore_features = ['Alley','PoolQC','MiscFeature','Fence','FireplaceQu','Utilities'],
             normalize = True,
             silent = True)
```

```python
lr = create_model('lr')
```
